# Análisis Exploratorio con D-Tale

**D-Tale** es una herramienta interactiva que combina las capacidades de pandas con una interfaz visual poderosa. Ofrece la posibilidad de explorar y analizar datos en tiempo real, permitiendo aplicar filtros, realizar transformaciones y generar gráficos sin necesidad de escribir código adicional.

Algunas funcionalidades clave de D-Tale son:
- Exploración y edición interactiva de datasets.
- Generación automática de gráficos personalizados.
- Estadísticas descriptivas para columnas individuales.
- Búsqueda rápida y agrupaciones dinámicas.

En este notebook, se empleó **D-Tale** para explorar de forma interactiva el dataset del Titanic, facilitando el descubrimiento de patrones y características mediante gráficos y estadísticas dinámicas.

```python
# Cargamos las librerías
import dtale
import pandas as pd
```

```python
# Cargamos dataset
df = pd.read_csv('../data/Titanic_Research_v6_clean.csv', sep= ';')

df.head()
```

```python
# Iniciar D-Tale
d = dtale.show(df)

# Abrir el navegador con el resultado
d.open_browser()

# También puedes generar una URL
# print(f"D-Tale está corriendo en: {d._main_url}")
```
